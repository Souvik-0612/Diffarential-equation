# 1. Diffarentiation

### Taylor series:
$$f(x+h) = f(x) + hf'(x)+\frac{h^2}{2!}f''(x)+\frac{h^3}{3!}f'''(x)+\frac{h^4}{4!}f''''(x)+...$$
## Backward diffarence:
Substitute $h$ by $-h$
$$f(x-h) = f(x) - hf'(x)+\frac{h^2}{2!}f''(x)-\frac{h^3}{3!}f'''(x)+\frac{h^4}{4!}f''''(x)+...$$
## Forward diffarence
Substitute $h$ by $h$
$$f(x+h) = f(x) + hf'(x)+\frac{h^2}{2!}f''(x)+\frac{h^3}{3!}f'''(x)+\frac{h^4}{4!}f''''(x)+...$$
## Central diffarence
Substact the forward and backward diffarence equations
$$f(x+h)-f(x-h) = 2\left(hf'(x)+\frac{h^3}{3!}f'''(x)+\frac{h^5}{5!}f'''''(x)+...\right)$$


### Note:
The main thing to note here is that if we use backward or forward diffarence then to find **diffarentiation** of a function using linear polynomial to fit (i.e, **straight line fit**) will have order of $h^2$ error.
Why? Incase of forward diffarence,

$$f'(x) = \frac{f(x+h) - f(x)}{h}$$

But in case of central diffarence we have $h^3$  error

$$f'(x) = \frac{f(x+h) - f(x-h)}{2h}$$

### Example1: 
Find the derivative of $\sin(x)$.

```python
#Importing all required libraries
import numpy as np
import matplotlib.pyplot as plt
plt.style.use(['science', 'notebook', 'grid'])
```

```python
#No of data point, you can change this to see the diffarence
n = 30

x, h = np.linspace(0, 4*np.pi, n, retstep = True) 
y = np.sin(x)
#The derivative of sin(x) is cos(x) which we get from analytically
yori = np.cos(x)

#To find the derivative of a function using forward diffarence
dF = []
for i in range(0, n-1):
    dF.append((y[i+1] - y[i])/h)

#To find the derivative of a function using backward diffarence
dB = []
for i in range(1, n):
    dB.append((y[i] - y[i-1])/h)
    
##To find the derivative of a function using central diffarence
dC = []
for i in range(1, n-1):
    dC.append((y[i+1] - y[i-1])/(2*h))
    
print(" Value of the diffarence ",h)

#plotting
plt.figure(figsize = (15, 8))
plt.title("Comparision between the various numerical methods for finding derivative")
plt.plot(x, yori, lw = 1,label = "Derivative")
plt.plot(x[0:n-1], dF, 'o', label = "Forward ")  #Slicing
plt.plot(x[1: n], dB, '*',lw = 1, label = "Backward")
plt.plot(x[1: n-1], dC, "-^", lw = 1, label = "Central")
plt.legend(loc = "best")
plt.show()
```

#### Observations 
1. Forward diffarence is undershoted from the original derivate function in first half period(i.e, $0$ to $\pi$) and it is overshoted in the remaining half period(i.e, $\pi$ to $2\pi$).
2. It is reversed in case of Backward diffarence.
3. But the central diffarence does great job!

We can estimate the error from the original deivative. To do that we have to take the absolute diffarence each element from the original array and the other arrays.

```python
#Errors between the Forward diffarence to original derivative
errF = abs(dF-yori[0: n-1]) #Broadcasting
errB = abs(dB-yori[1: n])
errC = abs(dC-yori[1: n-1])

#Plotting
plt.figure(figsize = (16, 8))
plt.title("Error analysis", size = 20)
plt.plot(x[0: n-1], errF, "o-", lw = 1, label = "Forward")
plt.plot(x[1: n], errB,"*-", lw = 1, label = "Backward")
plt.plot(x[1: n-1], errC, "^-",lw = 1, label = "Central")
plt.legend(loc = "best")
plt.show()
```
